```python
import glob
import cv2
import os

PREFIX_ORIGINAL = '/home/piggi/Desktop/ImageProcessed/Original'
PREFIX_FAKE = '/home/piggi/Desktop/ImageProcessed/Fake'

# Function to extract the file from a video and save this in a folder
def extract_frame_from_video(path, file):
    i = 0
    
    cap = cv2.VideoCapture(file)
    
    length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    while(cap.isOpened()):
        ret, frame = cap.read()
        i = i+1
        file_save = path +'/'+str(i)+'.jpg'
        cv2.imwrite(file_save,frame)
        if i == length:
            break
            
    cap.release()


# Read the file from the fake video folder and the original video one
print("Start to read the fake Folder with iglob:")
for fake_video in glob.iglob('/home/piggi/Desktop/deepfake/VidTIMIT/*'): 
    # Reaad the internal file in the folder
    # get the name of the folder where the file is present
    _path = os.path.basename(os.path.normpath(fake_video))
    #generate the path for the video file
    path = PREFIX_FAKE + "/" + _path
    try:
        os.makedirs(path)
    except OSError:
        print ("Creation of the directory %s failed" % path)
      
    for data in glob.iglob(fake_video+'/*'):
        #call the function for extract the file
        extract_frame_from_video(path,data)
print("End to read the Fake folder")  
    
print("Start to read the original Folder with iglob:")
for original_video in glob.iglob('/home/piggi/Desktop/deepfake/DeepfakeTIMIT/higher_quality/*'):
    # Read the internal file in the folder
    # get the name of the folder where the file is present
    _path = os.path.basename(os.path.normpath(original_video))
    #generate the path for the video file
    path = PREFIX_ORIGINAL + "/" + _path
    try:
        os.makedirs(path)
    except OSError:
        print ("Creation of the directory %s failed" % path)
        
    for data in glob.iglob(original_video+'/*'):
        #call the function for extract the file
        extract_frame_from_video(path,data)

print("End to read the Original Folder")

        

    
```

```python
# import the necessary packages
from skimage.measure import compare_ssim
import imutils
import cv2

# load the two input images
imageA = cv2.imread('/home/piggi/Desktop/ImageProcessed/Original/fcmh0/1.jpg')
imageB = cv2.imread('/home/piggi/Desktop/ImageProcessed/Fake/fcmh0/1.jpg')

#print(imageA)
#print('#######################')
#print(imageB)

# convert the images to grayscale
grayA = cv2.cvtColor(imageA, cv2.COLOR_BGR2GRAY)
grayB = cv2.cvtColor(imageB, cv2.COLOR_BGR2GRAY)

# compute the Structural Similarity Index (SSIM) between the two
# images, ensuring that the difference image is returned
(score, diff) = compare_ssim(grayA, grayB, full=True)
diff = (diff * 255).astype("uint8")
print("SSIM: {}".format(score))


# threshold the difference image, followed by finding contours to
# obtain the regions of the two input images that differ
thresh = cv2.threshold(diff, 0, 255,cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
```

```python
# Plot the histogram of the image

# importing required libraries of opencv 
import cv2 

# importing library for plotting 
from matplotlib import pyplot as plt 

# reads an input image 
original = cv2.imread('/home/piggi/Desktop/ImageProcessed/Original/fcmh0/1.jpg') 
fake = cv2.imread('/home/piggi/Desktop/ImageProcessed/Fake/fcmh0/1.jpg') 

# find frequency of pixels in range 0-255 
original_histr = cv2.calcHist([original],[0],None,[256],[0,256])
fake_histr = cv2.calcHist([fake],[0],None,[256],[0,256])

# show the plotting graph of an image 
plt.plot(original_histr)
plt.plot(fake_histr)
plt.show() 
```

```python
from mtcnn.mtcnn import MTCNN
import cv2

# Image filename
image_filename = '/home/piggi/Desktop/ImageProcessed/Original/fcmh0/1.jpg'

# Read the image file as a pixel array
img = cv2.imread(image_filename)

# Resize the image by half
width = int(img.shape[1]/2)
height = int(img.shape[0]/2)
img = cv2.resize(img, (width, height), interpolation = cv2.INTER_AREA)

# Display the image
#cv2.imshow('first', img)

# Convert the image to RGB from BGR format which Open CV uses as default
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

face_locations = []

mtcnn_face_detector_model = MTCNN()
face_locations = mtcnn_face_detector_model.detect_faces(rgb_image)
print('# of faces detected:', len(face_locations))

for face_location in face_locations:
  x1, y1, width, height = face_location['box']
  x2, y2 = x1 + width, y1 + height

  cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

cv2.imshow('first', img)

cv2.waitKey(0) # waits until a key is pressed
cv2.destroyAllWindows() # destroys the window showing image
```
